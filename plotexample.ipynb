# Darstellen von Punkten

## Darstellen von Punkten mit matplotlib 

```python
import pandas as pd

data = pd.DataFrame({"x": [11, 37, 98, 82], "y": [8, 5, 85, 95]})

pltimage = data.plot.scatter(x="x", y="y")
pltimage.set_xlabel("Klassenarbeit")
pltimage.set_ylabel("Durchschn. K1 bis K7")
pltimage.set_title("Übung 2")
```

## Darstellen von Punkten mit padas

```python
import pandas as pd

data = pd.DataFrame({"x":[1,2,3,4,5,6],"y":[6,5,4,3,2,1]})

pltimage = data.plot.scatter(x="x",y="y")
pltimage
```

# Zeichnen eines Tortendiagramms

## Tortendiagramm mit matplotlib

```python
import matplotlib.pyplot as plt

data = [1,2,3,4,5,6]
plt.pie(data,labels=["1","2",",3","4","5","6"])
plt.show()
```

## Tortendiagramm mit pandas

```python
import pandas as pd

data = pd.DataFrame([1,2,3,4,5,6])
pieimage = data.plot.pie(subplots=True)
pieimage
```
