```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV-Datei laden
data = pd.read_csv('Ladestationen.csv')


# Diagramm erstellen
plt.scatter(data['x'], data['y'], color='grey')

# Achsenbeschriftungen hinzufügen
plt.xlabel('X-Pos')
plt.ylabel('y-Pos')

#
# Diagramm anzeigen
plt.show()
```

Bestimmen geeigneter Positionen für die Ladestationen

```python
l_station=[[300,100],[650,300],[350,450]]

# Diagramm erstellen
plt.scatter(data['x'], data['y'], color='grey')
plt.scatter(l_station[0][0],l_station[0][1], s=80, color="red")
plt.scatter(l_station[1][0],l_station[1][1], s=80, color="blue")
plt.scatter(l_station[2][0],l_station[2][1], s=80, color="green")

# Achsenbeschriftungen hinzufügen
plt.xlabel('X-Pos')
plt.ylabel('y-Pos')

#
# Diagramm anzeigen
plt.show()
```

Einfärben der Stationen, die am dichtesten an der jeweiligen Ladestation sind

```python
import math

def distance(x1, y1, x2, y2):
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return dist

data_rot=[]
data_blau=[]
data_green=[]


for index,row in data.iterrows():
    
    d1=distance(l_station[0][0],l_station[0][1],row["x"],row["y"])
    d2=distance(l_station[1][0],l_station[1][1],row["x"],row["y"])
    d3=distance(l_station[2][0],l_station[2][1],row["x"],row["y"])
    
    dmin = min(d1,d2,d3)
    #print("dmin="+str(dmin)+" d1="+str(d1)+" d2="+str(d2)+" d3="+str(d3))
    if dmin==d1:
        data_rot.append([row["x"],row["y"]])
        #print("index="+str(index)+" appen to red")
    if dmin==d2:
        data_blau.append([row["x"],row["y"]])
        #print("index="+str(index)+" appen to blau")
    if dmin==d3:
        data_green.append([row["x"],row["y"]])
        #print("index="+str(index)+" appen to green")

plt.scatter([p[0] for p in data_rot], [p[1] for p in data_rot], color="red")
plt.scatter([p[0] for p in data_blau], [p[1] for p in data_blau], color="blue")
plt.scatter([p[0] for p in data_green], [p[1] for p in data_green], color="green")
plt.scatter(l_station[0][0],l_station[0][1], s=80, color="red")
plt.scatter(l_station[1][0],l_station[1][1], s=80, color="blue")
plt.scatter(l_station[2][0],l_station[2][1], s=80, color="green")

# Achsenbeschriftungen hinzufügen
plt.xlabel('X-Pos')
plt.ylabel('y-Pos')

#
# Diagramm anzeigen
plt.show()
```

Neue Ladestationen in das geometrische Mittel der Mengen setzen

```python
xm =0
ym = 0
for d in data_rot:
    xm=xm+d[0]
    ym=ym+d[1]

xm=xm/len(data_rot)
ym=ym/len(data_rot)
l_station[0]=[xm,ym]
xm =0
ym = 0
for d in data_blau:
    xm=xm+d[0]
    ym=ym+d[1]

xm=xm/len(data_blau)
ym=ym/len(data_blau)
l_station[1]=[xm,ym]
xm =0
ym = 0
for d in data_green:
    xm=xm+d[0]
    ym=ym+d[1]

xm=xm/len(data_green)
ym=ym/len(data_green)
l_station[2]=[xm,ym]

print("Neue Position der Ladestationen "+str(l_station))
```

Verwendung K-Mean Clustering aus sklearn

```python
import numpy as np
from sklearn.cluster import KMeans

# Definiere die Anzahl der Cluster
k = 3

# Erstelle ein KMeans-Objekt mit der Anzahl der Cluster
kmeans = KMeans(n_clusters=k)

# Wende den K-Means-Algorithmus auf die Daten an
kmeans.fit(data)

# Ermittle die Positionen der Zentroiden (d.h. der Punkte in der Mitte jedes Clusters)
centroids = kmeans.cluster_centers_

# Gib die Positionen der Zentroiden aus
print("Die Zentroiden sind:")


l_station[2]=centroids[2]
l_station[0]=centroids[1]
l_station[1]=centroids[0]
l_station
```
